```python
import turtle
import math

colors = ('Red','DarkOrange','Yellow','Lime','Cyan','Blue','Fuchsia','MediumBlue','DeepPink','OrangeRed','Navy','Green','Purple','Gray','SlateGray','Brown','SandyBrown','RosyBrown','SpringGreen','NavajoWhite','DarkCyan','Crimson','Salmon','HotPink','Violet','DarkViolet','RoyalBlue','Peru','DarkGoldenrod','MediumSpringGreen')

def triangle(t, length):
    t.pencolor('white') #Farbe der Umrandungslinie
    color = random.choice(colors) #Verbindung zu Liste 'colors'
    t.fillcolor(color) #wahl einer zufälligen Farbe
    t.begin_fill() #beginn Füllprozess
    for i in range(3): #durchführung für 3 Seiten
        t.fd(length) #Vorwärtsbewegung um angegebene Länge um Seite des Dreiecks zu zeichnen
        t.lt(120) #360/seitenzahl führt zu Winkel für ein gleichseitiges Dreieck    
    t.end_fill() #beendet den Füllprozess
    t.color('red') #Farbe der Schildkröte
    t.shape('turtle') #Abbildung des Pfeil wird durch Schildkröte ersetzt um den Niedlichkeitsfaktor zu erhöhen
        
def pile_triangle(t,length,x) :
    for i in range (x): #Anzahl der Dreiecke
        triangle(t,length) #Übergabe der Funktion Triangle
        t.lt(90) #die Folgenden drei Punkte bilden die Neuausrichtung, damit die Dreiecke nicht übereinander sondern nebeneinander gezeichnet werden
        t.fd(math.sqrt(3)*(length/2)) #bewegung um höhe des Dreiecks
        t.rt(90)

jimmy = turtle.Turtle()
pile_triangle(jimmy, 50, 5)
turtle.mainloop()
```

- **Aufgabe 9:** Schreiben Sie eine Funktion `tiled_triangle`, die **n Dreiecke übereinander und m Dreiecke nebeneinander** zeichnet. Sie soll vier Argumente  `t`, `l`, `n` und `m` erhalten. `t` ist eine Schildkröte, `l`, `n` und `m` Ganzzahlen. Die Funktion soll die Schildkröte nutzen, um `n` gleichseitige Dreiecke mit Seitenlänge `l` übereinander und `m` gleichseitige Dreiecke mit Seitenlänge `l` nebeneinander zu zeichnen.  <br>

```python
import turtle
import math
import random

colors = ('Red','DarkOrange','Yellow','Lime','Cyan','Blue','Fuchsia','MediumBlue','DeepPink','OrangeRed','Navy','Green','Purple','Gray','SlateGray','Brown','SandyBrown','RosyBrown','SpringGreen','NavajoWhite','DarkCyan','Crimson','Salmon','HotPink','Violet','DarkViolet','RoyalBlue','Peru','DarkGoldenrod','MediumSpringGreen')

def triangle(t, length):
    t.color('LemonChiffon') #Farbe der Schildkröte
    t.shape('turtle') #Abbildung des Pfeil wird durch Schildkröte ersetzt um den Niedlichkeitsfaktor zu erhöhen
    t.pencolor('LemonChiffon') #Farbe der Umrandungslinie
    color = random.choice(colors) #wählt eine zufällige Farbe aus der Liste colors aus
    t.fillcolor(color) #Bestimmung Farbe
    t.begin_fill() #Beginn Füllprozess
    for i in range(3): #durchführung für 3 Seiten
        t.fd(length) #Vorwärtsbewegung um angegebene Länge um Seite des Dreiecks zu zeichnen
        t.lt(120) #360/seitenzahl führt zu Winkel für ein gleichseitiges Dreieck    
    t.end_fill() #beendet den Füllprozess
    
def pile_triangle(t,length,x) :
    for i in range (x): #Anzahl der Dreiecke
        hight = math.sqrt(3)*(length/2) #Höhe des Dreiecks
        triangle(t,length) #Übergabe der Funktion Triangle
        t.penup() #Erhebung vom Stift, damit an einem Anderen Punkt weitergezeichnet werden kann
        t.lt(90) #die Folgenden drei Punkte bilden die Neuausrichtung, damit die Dreiecke nicht übereinander sondern nebeneinander gezeichnet werden
        t.fd(hight) #Neupositionierung um Höhe des Dreiecks
        t.rt(90)
        t.pendown() #Absetung des Stiftes
        
def repeat_triangle(t,length,x) :
    for i in range (x): #Anzahl der Dreiecke
        triangle(t,length) #Übergabe der Funktion Triangle
        t.rt(180) #die Folgenden drei Punkte bilden die Neuausrichtung, damit die Dreiecke nicht übereinander sondern nebeneinander gezeichnet werden
        t.fd(length)
        t.lt(180)
        
def zwischenfuell(t,length,x,a):
    for i in range (2): #2-malige Durchführung
        t.pencolor('LemonChiffon') #Linienfarbe 
        hight = math.sqrt(3)*(length/2) #berechnung höhe des dreiecks
        t.fd(a*length) # die nächsten 4 Punkte bilden die Berechnung und Zeichnung eines Rechteckes, welches mit den Variablen 
        t.lt(90)
        t.fd(x*hight) 
        t.lt(90)

def tiled_triangle(t,length,x,a) : #x = anzahl der Dreiecke in triangle_pile also der Reihen , a = anzahl der wiederholungen bzw. Spalten 
    t.fillcolor('LemonChiffon') #Festlegung Füllfarbe
    t.begin_fill() #Beginn Füllprozess
    zwischenfuell(t,length,x,a) #Aufruf von Zwischenfüll
    t.end_fill() #Ende Füllprozess
    for i in range (a): #Durchführung a-mal , gibt Anzahl der vertikalen Spalten an
        hight = math.sqrt(3)*(length/2) #Berechnung Höhe
        pile_triangle(t,length,x) #Aufruf der Funktion pile_triangle
        t.fd(length) #Neupositionierung
        t.rt(90)
        t.fd(x*hight)
        t.lt(90)

        
pete = turtle.Turtle()
tiled_triangle(pete,80,5,4)
turtle.mainloop()
```

```python
#def creative():
#while (schneeflöckchen)
#if ()
#for (Sterne)
import turtle
import math

def nose (t):
    n=7 #startwert für n
    for i in range (3): #3-malige Wiederholung
        t.fillcolor('OrangeRed') #Festlegung Füllfarbe
        t.begin_fill() #Start Füllprozess
        t.circle(n) #Zeichnung eines kreises mit dem Radius 7 zum Start als Basis für die Karotte
        t.end_fill() #Beendigung Füllprozess
        n = n * 0.8 #Neue Wertfestlegung für n damit die Kreise immer kleiner werden
        t.penup() #Hebung des Stiftes
        t.lt(90) #Neupositionierung
        t.fd(n*1.6)
        t.rt(90)
        t.pendown() #Absetzung Stift
    
def body (t,x):
    t.fillcolor('white') #Festlegung Füllfarbe
    for i in range (3): #3-malige Durchführung damit drei Schneekugeln entstehen
        t.begin_fill() #Anfang Füllprozess
        t.circle(x) #Kreis mit radius der Variable x um Körper des Schneemanns darzustellen
        t.end_fill() #Ende Füllprozess
        x = x * 0.7 #Festlegung eines neuen x-Wertes
        t.penup() #Anhebung Stift
        t.lt(90) #Neupositionierung um 1.8 faches der Höhe des ersten schneeballs damit es so aussieht als liegen sie aufeinander
        t.fd(x*1.8)
        t.rt(90)
        t.pendown() #Absetzung Stift
    t.penup() 
    t.rt(90) #Positionierung zum Startpunkt der nächsten Funktion
    t.fd(x*0.4)
    t.lt(90)
    t.fd(15)
    t.pendown()
        
def eyes (t):
    t.fillcolor('black') #Festlegung Füllfarbe
    t.begin_fill() #Start Füllprozess
    t.circle(5) #Auge aus Kreis mit radius 5
    t.end_fill() #Ende Füllprozess
    t.penup() #Anhebung Stift
    t.lt(180) #Neupositionierung zum Punkt des zweiten Auges
    t.fd(30)
    t.lt(180)
    t.pendown() #Stift absetzen
    t.begin_fill() #zweiter Start des Füllprozesses da turtle sonst auch den bereich füllen will, bei dem der stift angehoben wird
    t.circle(5) #zweites auge
    t.end_fill()
    t.penup()
    t.fd(10) #Neupositionierung
    t.rt(90)
    t.fd(5)
    t.rt(35)
    t.pendown()
    
def buttons(t):
    t.penup() #Anhebung stift
    t.lt(35) #Neupositionierung
    t.fd(30)
    t.rt(90)
    t.fd(10)
    t.lt(90)
    t.pendown() #Absetzung stift
    t.fillcolor('black') #Festlegung Füllfarbe
    t.begin_fill() #Füllprozess
    t.circle(5) #erster Knopf
    t.end_fill()
    t.penup()
    t.fd(20)
    t.pendown()
    t.fillcolor('black')
    t.begin_fill()
    t.circle(5) #zweiter Knopf weiter unten
    t.end_fill()
    t.penup()
    t.fd(45)
    t.pendown()
    t.fillcolor('black')
    t.begin_fill()
    t.circle(5) #dritter Knopf ganz unten
    t.end_fill()

def hat (t):
    t.penup()
    t.lt(180) #Neupositionierung
    t.fd(141)
    t.lt(90)
    t.fd(40)
    t.pendown()
    t.pensize(10) #Änderung der Breite der Linie auf Größe 10
    t.lt(180) #Zeichnung der hutkrempe
    t.fd(76)
    t.lt(180)
    t.pensize(1)
    t.fillcolor('black')
    t.begin_fill()
    t.fd(60) #Schritte bis zu nächsten hashtag zeichen bilden das Rechteck des Hutes
    t.rt(90)
    t.fd(60)
    t.rt(90)
    t.fd(40)
    t.rt(90)
    t.fd(60)
    t.rt(90)
    t.fd(40)#
    t.rt(180) #für richtige Porsition des Baumes
    t.end_fill()
    
def snowman(t,x): #Aufruf aller Schneemann Fuktionen und Kombination zu komplettem Schneeman
    t.pencolor('black')
    t.penup()
    t.goto(-120,-100)
    t.lt(47)
    t.pendown()
    body(t,x)
    eyes(t)
    nose(t)
    buttons(t)
    hat(t)
    
        
def snowglobe(t,x): 
    t.penup() #Positionierung
    t.goto(0,0)
    t.rt(90)
    t.fd(x*3)
    t.lt(90)
    t.pendown()
    t.fillcolor('WhiteSmoke') #Festlegung Füllfarbe
    t.begin_fill() #Zeichnung eines großen Kreises für die Schneekugel
    t.circle(x*4)
    t.end_fill()
    
def fallen_snow(t):
    t.fillcolor('white')
    t.begin_fill() 
    t.penup() #Neupositionierung
    t.goto(-198,-127)
    t.pendown()
    t.seth(-45) #dreht turtle passend zur Eingabe
    t.pencolor('white')
    t.circle(279,90) #es werden im Prinzip 4 Kreise gezeichnet, jedoch von diesen Kreisen werden nur 90 Grad gezeichnet, sodass man ein Oval oder eine andere Abgerundete Form zeichnen kann
    t.pencolor('black')
    t.circle(53,90)
    t.circle(282,90)
    t.circle(52,90)
    t.end_fill()

def triangle(t,length): #Grundlage für den Weihnachtsbaum
    t.fillcolor('DarkGreen')
    t.begin_fill()
    for i in range (3): #durchführung für 3 Seiten
        t.fd(length) #Vorwärtsbewegung um angegebene Länge um Seite des Dreiecks zu zeichnen
        t.lt(120) #360/seitenzahl führt zu Winkel für ein gleichseitiges Dreieck  
    t.end_fill() #beendet den Füllprozess
    
def star (t):
    i = 1 #startwert für i
    t.pencolor('gold') #Änderung Linienfarbe
    t.fillcolor('gold') #Anderung Füllfarbe
    t.begin_fill()
    t.lt(36) #Startpositionierung
    while i in range (5) : #Einbindung einer While Schleife, welche Verursacht dass der Prozess für die 5 Striche die für ein Pentagram benötigt werden
        t.lt(144) #Gradzahl benötigt für einen gleichmäßigen stern
        t.fd(50) #Strichlänge
        i = i + 1 #Erhöhung von i um 1
    t.end_fill()

def tree (t,length):
    i = 1
    t.penup()
    t.goto(65,-70)    
    t.pendown()
    t.fillcolor('SaddleBrown')
    t.begin_fill()
    for i in range (4): #Rechteck für Baumstamm
        t.fd(40)
        t.lt(90)
    t.end_fill()
    t.penup() #Neupositionierung
    t.goto(-15,-50)
    t.pendown()
    for i in range (3): #Dreimaliger Aufruf der Dreiecks Funktion wobei die Dreiecke bei jedem Durchlauf kleiner werden und nach oben gesetzt werden
        triangle(t,length)
        t.penup()
        t.fd(length/2)
        t.lt(90)
        t.fd(length*0.4)
        t.lt(90)
        length = length * 0.8
        t.fd(length/2)
        t.rt(180)
        i = i + 1 
        t.pendown()
    t.penup()
    t.goto(100,230) #Neupositionierung für Stern an der Spitze
    t.pendown()
    star(t) #Aufruf von Stern
    
def socket(t): #Trapez, welches teils von der <kugel überdeckt wird
    t.penup() #Positionierung
    t.goto(-197,-80)
    t.pendown()
    t.fillcolor('Maroon') #Festlegung Füllfarbe
    t.begin_fill()
    t.fd(400) #Zeichnung des Trapezes
    t.rt(80)
    t.fd(200)
    t.rt(100)
    t.fd(480)
    t.rt(100)
    t.fd(200)
    t.rt(80)
    t.end_fill()
    
def snowflake_arm (t): #Einzelteil der Schneeflocke, welche wiederholt wird
    for i in range(3): #Astbildung  und Zeichnung der Grundstruktur der Schneeflocke
        for i in range(3):
            t.fd(5)
            t.lt(180)
            t.fd(5)
            t.lt(180)
            t.rt(45)
        t.lt(90)
        t.lt(180)
        t.fd(5)
        t.lt(180)
        t.lt(45)
    t.rt(90) #Neupositionierung
    t.fd(15)

def snowflake (t):
    t.pencolor('black') #Festlegung Linienfarbe
    t.speed(500) #Beschleunigung, da der Prozess sonst ewig lange deuern würde
    for i in range(8): #Wiederholung der Funktion snowflake_arm 8 mal
        snowflake_arm(t)
        t.lt(45) #Neupositionierung
    
def falling_snow (t) : #Abwechselnde Zeichnung von Schneeflocken als snowflake und als weiß gefüllter Kreis an verschiedenen Koordinaten
    t.penup()
    t.goto(30,70)
    t.pendown()
    snowflake(t)
    t.penup()
    t.goto(-20,0)
    t.pendown()
    t.fillcolor('white')
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    t.goto(100,-80)
    t.pendown()
    snowflake(t)
    t.penup()
    t.goto(180,-100)
    t.pendown()
    t.begin_fill()
    t.circle(3)
    t.end_fill
    t.penup()
    t.goto(200,50)
    t.pendown()
    snowflake(t)
    t.penup()
    t.goto(130,160)
    t.pendown()
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    t.goto(-160,150)
    t.pendown()
    snowflake(t)
    t.begin_fill()
    t.penup()
    t.goto(-110,-100)
    t.pendown()
    snowflake(t)
    t.penup()
    t.goto(-190,40)
    t.pendown()
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    t.goto(-20,-190)
    t.pendown()
    snowflake(t)
    t.penup()
    t.goto(-80,190)
    t.pendown()
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    t.goto(-10,230)
    t.pendown()
    snowflake(t)

def shake (t): 
    answer = input("Willst du die Schneekugel schütteln? Antworte ja / nein. ")
    if answer == "ja": #implementierung einer if verzweigung welche auf der Eingabe einer Anwort zur Ausgegebenen Frage beruht
        falling_snow(t) #Aufruf von falling_snow
    else:
        print("Schade...") #Ausgabe des in den '' stehenden textes

#!!!!!!!!!!!!!!!!
#Link des im Hintergrund zu spielenden Liedes: https://www.youtube.com/watch?v=gm3puQtwXcg&list=PLor7HEPF5NWchDXQtb2gjw3sf9QAFMuEa&index=2&t=63s
#!!!!!!!!!!!!!!!!
santa = turtle.Turtle()
socket(santa)
snowglobe(santa,70)
fallen_snow(santa)
snowman(santa,70)
tree(santa,200)
shake(santa)
turtle.mainloop()
```
